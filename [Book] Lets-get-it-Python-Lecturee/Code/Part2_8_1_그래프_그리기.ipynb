```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
plt.plot([1, 5, 7, 3, 7])
plt.show()
```

- plot() 명령어는 데이터가 단일 값이면 X축은 0부터 시작, Y축은 입력된 데이터를 나타낸다

```python
# X, Y 데이터로 그래프 그리기
month = [3, 4, 5, 6, 7]
sales = [1, 5, 7, 3, 7]
plt.plot(month, sales)
plt.show()
```

```python
month = ['mar', 'apr', 'may', 'jun', 'jul']
sales = [1, 5, 7, 3, 7]
plt.plot(month, sales)
plt.show()
```

- X축 데이터와 Y축 데이터는 항상 개수가 동일해야 한다.
    - 다를 경우 오류 발생

```python
# 제목 붙이기
month = ['mar', 'apr', 'may', 'jun', 'jul']
sales = [1, 5, 7, 3, 7]
plt.title('월별 판매 실적') # 그래프 제목 넣기
plt.plot(month, sales)
plt.show()
```

```python
# 그래프 색 바꾸기
plt.title('월별 판매 실적') # 그래프 제목 넣기
plt.plot(month, sales, color='r')
plt.show()
```

```python
# 크기 바꾸기(dpi)
plt.figure(dpi=150)
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b')
plt.show()
```

```python
# 크기 바꾸기(figsize)
plt.figure(figsize=(8, 4))
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b')
plt.show()
```

```python
# 범례 넣기
plt.figure(dpi=100)
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위: 개수') # 범례 넣기
plt.legend()
plt.show()
```

```python
# 범례 위치 변경
plt.figure(dpi=100)
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위: 개수') # 범례 넣기
plt.legend(loc='lower right')
plt.show()
```

```python
# 축 이름 넣기
plt.figure(dpi=100)
plt.title('월별 판매 실적')
plt.plot(month, sales, color='b', label='단위: 개수')
plt.legend(loc='lower right')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
