```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')
```

```python
# 막대그래프 그리기
sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]
plt.bar(month, sales)
plt.show()
```

```python
# 제목 삽입 및 그래프 색 바꾸기
sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]
plt.title('월별 판매 실적')
plt.bar(month, sales, color='b')
plt.show()
```

```python
# 범례 넣기
sales = [1, 5, 7, 3, 7]
month = [3, 4, 5, 6, 7]
plt.title('월별 판매 실적')
plt.bar(month, sales, color='b', label='단위: 개수')
plt.legend()
plt.show()
```

```python
# x축에 range()로 연속 숫자 넣기
sales = [1, 5, 7, 3, 7]
plt.title('월별 판매 실적')
plt.bar(range(3, 8), sales, color='b', label='단위: 개수')
plt.show()
```

```python
sales = [1, 5, 7, 3, 7]
plt.title('월별 판매 실적')
plt.bar(range(1, 6), sales, color='r', label='단위: 개수')
plt.show()
```

```python
sales = [1, 5, 7, 3, 7]
plt.title('월별 판매 실적')
plt.bar(range(2, 11, 2), sales, color='g', label='단위: 개수')
plt.show()
```

```python
# 막대 방향 전환하기
sales = [1, 5, 7, 3, 7]
plt.title('월별 판매 실적')
plt.barh(range(2, 11, 2), sales, color='r', label='단위: 개수')
plt.show()
```

```python
# 히스토그램 그리기
sales = [1, 5, 7, 3, 7]
plt.hist(sales)
plt.show()
```

```python
import random

a = []
for i in range(10):
    a.append(random.randint(1, 10))
print(a)
plt.hist(a)
plt.show()
```

```python
# 원그래프 그리기
b_type = [25, 19, 37, 11]
plt.pie(b_type)
plt.show()
```

```python
# 라벨 표시하기
b_type = [25, 19, 37, 11]
plt.pie(b_type, labels=['A형', 'B형', 'C형', 'AB형'])
plt.show()
```

```python
b_type = [25, 19, 37, 11]
b_name = ['A형', 'B형', 'C형', 'AB형']
plt.pie(b_type, labels=b_name)
plt.show()
```

```python
# 비율 표시하기
b_type = [25, 19, 37, 11]
b_name = ['A형', 'B형', 'C형', 'AB형']
plt.pie(b_type, labels=b_name, autopct='%1.1f%%')
plt.show()
```

```python
b_type = [25, 19, 37, 11]
b_name = ['A형', 'B형', 'C형', 'AB형']
plt.pie(b_type, labels=b_name, autopct='%1.2f%%') # 소수점 2번째자리
plt.show()
```

```python
# 산점도 그리기
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
plt.scatter(kor, eng)
plt.show()
```

```python
# 색 바꾸고 투명도 설정하기
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
plt.scatter(kor, eng, color='red', alpha=0.7)
plt.show()
```

```python
kor = [80, 20, 50, 20, 10, 50, 60, 30, 60]
eng = [90, 40, 60, 40, 10, 30, 50, 70, 90]
plt.scatter(kor, eng, color='red', alpha=0.3)
plt.xlabel('국어 점수')
plt.ylabel('영어 점수')
plt.show()
```

```python
# 컬러맵 적용하기
height, weight = [], []
for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=height, cmap='RdPu')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=weight, cmap='BuPu')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
plt.title('키와 몸무게의 상관관계')
plt.scatter(height, weight, c=height, cmap='Oranges')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='키')
plt.show()
```

```python
# 점의 크기 조절하기
height, weight, score = [], [], []
for i in range(100):
    height.append(random.randint(100, 200))
    weight.append(random.randint(20, 100))
    score.append(random.randint(1, 1000))

plt.title('키와 몸무게, 체육 점수의 상관관계')
plt.scatter(height, weight, s=score, c=score, cmap='rainbow', alpha=0.7)
plt.xlabel('키')
plt.ylabel('몸무게')
plt.colorbar(label='체육 점수')
plt.show()
```
