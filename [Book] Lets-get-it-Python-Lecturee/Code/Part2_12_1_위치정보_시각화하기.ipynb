### 12. 위치정보 시각화하기: folium

#### 12.1.1 folium 설치하고 지도 만들기

```python
import folium
```

- folium.Map([위도, 경도])

```python
# '도산공원' 위치 표시
map_dosan = folium.Map([37.524664222356904, 127.03534524001842])
map_dosan
```

```python
# 지도 비율 조정하기
map_dosan = folium.Map([37.524664222356904, 127.03534524001842], zoom_start=15)
map_dosan
```

```python
# 위도, 경도 변수 활용하여 나타내기
lat, long = 37.524664222356904, 127.03534524001842
map_dosan = folium.Map([lat, long], zoom_start=20)
map_dosan
```

```python
# 마커 추가하기
lat, long = 37.524664222356904, 127.03534524001842
map_dosan = folium.Map([lat, long], zoom_start=17)
folium.Marker([lat, long]).add_to(map_dosan)
map_dosan
```

```python
# 여러 장소 마커 표시하기
lat1, long1 = 37.524664222356904, 127.03534524001842 # 도산공원
lat2, long2 = 37.52416006563674, 127.03191201275821 # 신사무궁화공원

map_dosan = folium.Map([lat1, long1], zoom_start=17)

folium.Marker([lat1, long1]).add_to(map_dosan)
folium.Marker([lat2, long2]).add_to(map_dosan)
map_dosan
```

```python
# 마커에 텍스트 표시하기
# tooltip은 마커에 마우스 올리면 텍스트 표시, 벗어날 시 미 표시
# popup은 마커 클릭 시 말풍선 모양으로 표시, 재클릭 시 사라짐

lat1, long1 = 37.524664222356904, 127.03534524001842 # 도산공원
lat2, long2 = 37.52416006563674, 127.03191201275821 # 신사무궁화공원

map_dosan = folium.Map([lat1, long1], zoom_start=17)

folium.Marker([lat1, long1], tooltip='도산 공원').add_to(map_dosan)
folium.Marker([lat2, long2], popup='신사무궁화공원').add_to(map_dosan)
map_dosan
```

```python
# 마커의 아이콘과 색 변경하기
lat1, long1 = 37.524664222356904, 127.03534524001842 # 도산공원
lat2, long2 = 37.52416006563674, 127.03191201275821 # 신사무궁화공원

map_dosan = folium.Map([lat1, long1], zoom_start=17)

folium.Marker([lat1, long1], tooltip='도산 공원', icon=folium.Icon(color='red')).add_to(map_dosan)
folium.Marker([lat2, long2], tooltip='신사무궁화공원', icon=folium.Icon(color='purple')).add_to(map_dosan)
map_dosan
```

```python
# 마커 모양 변경하기
lat1, long1 = 37.524664222356904, 127.03534524001842 # 도산공원
lat2, long2 = 37.52416006563674, 127.03191201275821 # 신사무궁화공원

map_dosan = folium.Map([lat1, long1], zoom_start=17)

folium.Marker([lat1, long1], tooltip='도산 공원', icon=folium.Icon(color='red', icon='heart')).add_to(map_dosan)
folium.Marker([lat2, long2], tooltip='신사무궁화공원', icon=folium.Icon(color='purple', icon='home')).add_to(map_dosan)
map_dosan
```

```python
# 반복문 활용하여 위치 표시하기
lat = [37.524664222356904, 37.52416006563674]
long = [127.03534524001842, 127.03191201275821]
names = ['도산공원', '신무궁화공원']
icons = ['heart', 'home']
colors = ['red', 'purple']

map_dosan = folium.Map([lat[0], long[0]], zoom_start=16)

for i in range(len(lat)):
    folium.Marker(
                [lat[i], long[i]], tooltip=names[i],
                icon=folium.Icon(color=colors[i], icon=icons[i])
    ).add_to(map_dosan)
    
map_dosan
```
