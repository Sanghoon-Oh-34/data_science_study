```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

import csv
```

```python
# 파일 열기
f = open('/content/drive/MyDrive/Colab Notebooks/Lets_get_it/card.csv')

# 파일 읽기
data = csv.reader(f)

# 헤더 제거(첫 번째행 건너띄기)
next(data)

# 리스트로 변환
data = list(data)

print(data)
```

### 9.2 데이터 추출: 필요한 데이터만 골라내기

```python
# 데이터 타입 확인하기
print(type(data))
```

```python
# 데이터 길이 구하기(2차원 리스트이기때문에 길이는 행의 개수를 의미)
print(len(data))
```

```python
# data 여러 행 출력하기
print(data[:3])
```

```python
# 특정 값 골라서 출력하기(첫 행의 이용내역)
print(data[0][6])
```

```python
# 첫 행의 가맹점명과 이용금액 출력하기
print(data[0][5], data[0][6])
```

```python
# 첫 행 - 이용일시, 가맹점명, 이용금액 출력하기
print(data[0][0], data[0][5], data[0][6])
```

```python
# 반복문으로 데이터 추출하기
for row in data:
    print(row)
```

```python
# 이용금액만 출력
for row in data:
    print(row[6])
```

```python
# 반복문 - 이용일시, 가맹점명, 이용금액 출력하기
for row in data:
    print(row[0], row[5], row[6])
```

- CSV 파일에서 읽어 들인 데이터는 모두 문자열로 인식된다

```python
# 이용금액 타입변환(str > int)
for row in data:
    payment = int(row[6])
    print(payment, type(payment))
```

```python
# 가맹점명, 이용금액 출력하기(변수 저장)
for row in data:
    payment = int(row[6])
    store = row[5]
    print(store, '에서', payment, '원 결제')
```

```python
# 이용일시, 가맹점명, 이용금액 출력하기(날짜 형식 변경)
for row in data:
    date = row[0].split()[0]
    store = row[5]
    payment = int(row[6])
    print(date, store, '에서', payment, '원 결제')
```

```python
# 이용내역에 대한 조건문 출력하기
for row in data:
    if row[-1] == '전표매입':
        print('구매확정!')
    else:
        print('구매취소!')
```

```python
# 총 지출액 구하기
spend = 0
for row in data:
    if row[-1] == '전표매입':
        spend += int(row[-3])
print(spend)
```

```python
# 월별 지출액 그래프 그리기
s_mon = [0, 0, 0]
for row in data:
    if row[-1] == '전표매입':
        mon = int(row[0].split('-')[1])
        payment = int(row[-3])
        idx = mon - 10
        s_mon[idx] += payment
print(s_mon)
```

```python
s_mon = [0, 0, 0]
for row in data:
    if row[-1] == '전표매입':
        s_mon[int(row[0].split('-')[1]) - 10] += int(row[-3])
print(s_mon)
```

```python
plt.title('10~12월 지출현황')
plt.bar(['10월', '11월', '12월'], s_mon, color='royalblue')
plt.show()
```

```python
# 월별 택시비 지출액 그래프 그리기
taxi = [0, 0, 0]
for row in data:
    if row[-1] == '전표매입' and '택시' in row[5]:
        taxi[int(row[0].split('-')[1]) - 10] += int(row[-3])
print(taxi)
```

```python
plt.title('10~12월 택시비 지출현황')
plt.plot(['10월', '11월', '12월'], taxi, color='crimson', label='택시비 지출액')
plt.legend()
plt.show()
```

```python
# 월별 배달음식비 지출액 그래프 그리기
deli = [0, 0, 0]
for row in data:
    if row[-1] == '전표매입' and row[5] == '(주)우아한형제들':
        deli[int(row[0].split('-')[1]) - 10] += int(row[-3])
print(deli)
```

```python
plt.title('10~12월 배달음식비 지출현황')
plt.plot(['10월', '11월', '12월'], deli, color='indigo', label='배달음식비 지출액')
plt.legend()
plt.show()
```

```python
# 비교 그래프 합치기
plt.title('10~12월 택시비/배달음식비 지출현황')
plt.plot(['10월', '11월', '12월'], taxi, color='crimson', label='택시비 지출액')
plt.plot(['10월', '11월', '12월'], deli, color='indigo', label='배달음식비 지출액')
plt.legend()
plt.show()
```

```python
# 지출액 상위 10개 항목 뽑기
spending = {}
for row in data:
    if row[-1] == '전표매입':
        store, payment = row[-4], int(row[-3])
        if store not in spending.keys():
            spending[store] = payment
        else:
            spending[store] += payment

print(spending)
```

- 딕셔너리 Value를 이용한 정렬
    - sorted()와 operator를 이용

```python
import operator
```

```python
top10 = sorted(spending.items(), key=operator.itemgetter(1), reverse=True)[:10]
print(top10)
```

```python
# 지출액 상위 10개의 가맹점명과 지출액을 리스트에 저장하기
top10_store, top10_amount = [], []
for t in top10:
    top10_store.append(t[0])
    top10_amount.append(t[1])
```

```python
print(top10_store)
```

```python
print(top10_amount)
```

```python
# 가맹점명, 지출액 막대그래프 그리기
plt.title('10~12월 지출 TOP10')
plt.barh(top10_store, top10_amount, color='b')
plt.show()
```
